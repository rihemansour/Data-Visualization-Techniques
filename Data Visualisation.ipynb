### Line Plot
A line plot, also known as a line chart or line graph, is a type of data visualization that displays data points connected by straight lines. It is typically used to show the relationship between two continuous variables over a continuous interval or time period. Line plots are particularly useful for displaying trends and patterns in data, making them suitable for time series data or any data where you want to show how a variable changes over time.

**Data Type:** Line plots are best suited for continuous data, such as numerical data points that can be measured on a continuous scale (e.g., time, temperature, sales figures).

**Example:** Let's say you want to visualize the monthly expenses a college student spends in Tunisia.



```python
months = ["Sep", "Oct", "Nov", "Dec", "Jan", "Feb", "Mar", "Apr", "May", "Jun"] 
expenses = [124, 283, 192, 221, 305, 176, 132, 297, 314, 189]
```

```python
fig , ax = plt.subplots()
ax.plot(months,expenses)
ax.set_xlabel("Months of the academic year")
ax.set_ylabel("Expenses of the student in Dt")
ax.set_title("Monthly expenses a college student spends in Tunisia")
ax.grid(True)
plt.show()
```



Suppose we want to visualize the expenses of two students to compare their spending; we can proceed like this :



```python
student_1 = [124, 283, 192, 221, 305, 176, 132, 297, 314, 189]
student_2 = [250, 275, 110, 230, 120, 340, 275, 123, 180, 300]

fig , ax = plt.subplots()
ax.plot(months, student_1, label="Student 1")
ax.plot(months, student_2, label="Student 2")
ax.set_xlabel("Months of the academic year")
ax.set_ylabel("Expenses of the student in Dt")
ax.set_title("Monthly expenses a college student spends in Tunisia")
ax.grid(True)
ax.legend()
plt.show()
```

### Scatter Plot
A scatter plot, also known as a scatter chart or scattergraph, is a type of data visualization used to display the relationship between two variables. It is particularly useful for showing how one variable is affected by another and for identifying patterns, trends, or correlations in data.

***Data Type:*** Scatter plots are used for displaying two continuous variables, where each data point represents a unique combination of values for the two variables. These variables are often referred to as the independent variable (usually on the x-axis) and the dependent variable (usually on the y-axis).

***Example:*** Let's consider the relationship between the number of hours of exercise and the number of calories burned:



```python
hours_of_exercise = [1, 2, 2.5, 3, 3.5, 4, 4.5, 5]
calories_burned = [150, 300, 375, 450, 525, 600, 675, 750]

plt.scatter(hours_of_exercise,calories_burned)
plt.xlabel("Hours of exercise")
plt.ylabel("Number of calories burned")
plt.title("Relationship between the number of hours of exercise and the number of calories burned")
plt.show()
```

### Scatter matrix plot
A scatter matrix plot, also known as a scatterplot matrix, is a data visualization technique used to explore the relationships between multiple variables (attributes or features) in a dataset. It consists of a grid of scatterplots where each pair of variables is plotted against each other. The diagonal of the grid usually contains histograms or density plots of the individual variables. Scatter matrix plots help identify patterns, correlations, and potential outliers in multivariate data.

***Data type:*** Scatter matrix plots are particularly useful when dealing with multivariate datasets, where you have several variables or attributes. Each variable should be continuous or numeric to create meaningful scatterplots.

***Example:*** The following example is excerpted from the renowned book 'Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow' by the author Aurlien Gron. It demonstrates how to assess correlations among attributes within the well-known housing pricing dataset.

```python
import os
import tarfile
import urllib
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    os.makedirs(housing_path, exist_ok=True)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv") 
    return pd.read_csv(csv_path)
fetch_housing_data()
housing = load_housing_data()
housing.head()
from pandas.plotting import scatter_matrix
attributes = ["median_house_value", "median_income", "total_rooms","housing_median_age"]
scatter_matrix(housing[attributes], figsize=(12, 8))
```

***Data Type:*** Bar plots are particularly suitable for displaying discrete or categorical data, where the data points fall into distinct categories or groups. The length or height of each bar represents a quantitative value associated with that category.

### Histogram
A histogram is a graphical representation of the distribution of a dataset. It divides the data into discrete bins or intervals and displays the frequency or count of data points falling into each bin using bars. Histograms are useful for visualizing the underlying data distribution and identifying patterns or characteristics within it.

***Data Type:*** Histograms are typically used for displaying continuous data, where the data points fall on a continuous scale, and you want to understand the frequency or distribution of values within that scale.
***Example:*** We'll create a histogram to visualize the distribution of ages in a sample population:

```python
import numpy as np
ages = np.random.normal(50,10,200)
plt.hist(ages,bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age Distribution")
plt.show()
```

### Pie chart
A pie chart is a circular data visualization that represents data as slices of a pie or circle. Each slice, or "pie wedge," represents a category or proportion of the whole dataset, and the size of each slice corresponds to the quantity or percentage of that category relative to the total.

***Data Type:*** Pie charts are used to display categorical data where you want to illustrate the proportions or distribution of categories within a whole. The values of the categories are typically expressed as percentages of the whole or as fractions of a circle.

***Example:***
Let's create a pie chart to visualize the distribution of a student's monthly expenses across different expense categories.

```python
expense_categories = ['Rent', 'Food', 'Transportation', 'Entertainment', 'Utilities']
expenses = [200, 120, 100, 150, 50]
plt.pie(expenses, labels=expense_categories,autopct="%1.1f%%")
plt.title('Monthly Expenses Distribution')
plt.show()
```

### Doughnut chart
A doughnut chart is a circular data visualization that is similar to a pie chart but with a hole in the center, creating a "doughnut" shape. It is used to represent data in the form of a ring, where each category or data point is represented by a segment of the ring's circumference. Doughnut charts are particularly useful for displaying part-to-whole relationships and emphasizing the composition of a whole while comparing the individual components.

***Data type:*** Doughnut charts are suitable for representing data with multiple categories or components, where each category's value contributes to the total. They are often used to show the distribution of a whole into its constituent parts, typically expressed as percentages or proportions.

***Example:*** we'll use the same example as before :

```python
plt.figure(figsize=(6, 6))
plt.pie(expenses,labels=expense_categories, autopct='%1.1f%%', wedgeprops={'edgecolor': 'white'})
# Add a circle in the center to create the "hole"
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.title('Monthly Expenses Distribution')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Box plot
Before we delve into the explanation of this plot, it's essential to recall the definition of quartiles.

***Quartiles*** are statistical measures that divide a data set into four equal parts, each containing approximately 25% of the data points. They are used to understand the distribution and spread of data in a dataset. Quartiles are often used in descriptive statistics and data analysis to summarize and compare different parts of a dataset.

There are three quartiles commonly used:

***First Quartile (Q1):*** This is the 25th percentile of the data, which means that 25% of the data points fall below this value. It is also known as the lower quartile.

***Second Quartile (Q2):*** This is the 50th percentile of the data, which is also the median. It divides the data into two equal halves, with 50% of the data points falling below and 50% above this value.

***Third Quartile (Q3):*** This is the 75th percentile of the data, which means that 75% of the data points fall below this value. It is also known as the upper quartile.

Quartiles are particularly useful for identifying outliers and assessing the spread of data. The interquartile range (IQR), which is the difference between the third quartile (Q3) and the first quartile (Q1), is often used to measure the spread or variability of data. It provides a robust measure of dispersion that is less sensitive to outliers than the range. Outliers, which are data points significantly different from the majority of the data, can be identified by comparing them to the quartiles and the IQR.

A ***box plot***, also known as a ***box-and-whisker plot***, is a graphical representation of the distribution of a dataset. It displays a summary of the dataset's central tendency, spread, and potential outliers. A box plot consists of a rectangular "box" that represents the interquartile range (IQR) of the data, and "whiskers" that extend from the box to show the data's range.

***Data Type:*** Box plots are used to visualize the distribution of numerical data, especially when comparing multiple datasets or identifying the presence of outliers.

Components of a Box Plot:


***Box:*** Represents the IQR (interquartile range), which contains the middle 50% of the data. The box spans from the lower quartile (Q1) to the upper quartile (Q3).

***Whiskers:*** Lines extending from the box to the minimum and maximum values within a certain range, usually 1.5 times the IQR.

***Outliers:*** Individual data points that fall outside the whiskers and are considered potential outliers.

***Example:***
Let's create a box plot to visualize the distribution of test scores for two different classes, Class A and Class B. Here's an example dataset:

```python
class_a_scores = [13, 7, 2, 14, 14, 12, 13, 19, 20, 1]
class_b_scores = [19,17,18,20,17,16,19,18,19,18,2,1,3]
class_c_scores = [4, 2, 6, 2, 1, 18, 5, 19, 2, 3]

plt.boxplot([class_a_scores,class_b_scores,class_c_scores],labels=["class_a_scores","class_b_scores","class_c_scores"])
plt.show()
```

### Violin plot
A violin plot is a data visualization that combines elements of a box plot and a kernel density plot. It is used to display the distribution of data across different categories or groups. Violin plots typically consist of one or more "violins," which resemble violin shapes and represent the distribution of data within each category. The width of the violin at a given point represents the density of data at that value. Violin plots provide insights into the data's distribution, central tendency, and spread.

***Data type:*** Violin plots are particularly useful when you want to compare the distribution of data between multiple categories or groups. They are effective for visualizing continuous or numerical data within these categories.

***Example:*** We'll use the same example as before :

```python
plt.violinplot([class_a_scores,class_b_scores,class_c_scores])
plt.xticks([1,2,3],["class_a","class_b","class_c"])
plt.ylabel("scores")
plt.title("Violin Plot of Exam Scores by Class")
plt.show()
```

### Heatmap
A heatmap is a data visualization technique that uses colors to represent the values of a matrix or a 2D dataset. Heatmaps are often used to visualize relationships or patterns in data, especially when dealing with large datasets. In a heatmap, each cell in the matrix is typically assigned a color based on its value, with colors representing different levels or intensities. Heatmaps are useful for identifying clusters, trends, or correlations in data.They are commonly used in fields such as data analysis, biology (e.g., gene expression data), and business (e.g., sales data).

***Data type:*** Heatmaps are versatile and can be used with various types of data, including numerical data, categorical data, and binary data. 

***Example:*** Visualizing user interaction on a website through a heatmap to optimize user experience.


```python
data = np.random.randn(4,4)
plt.imshow(data)
plt.colorbar()
plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns

correlation_matrix = np.corrcoef(data)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

### Area Plot
An area plot, also known as a stacked area chart, is a data visualization that displays multiple datasets as stacked filled areas, representing the cumulative contribution of each dataset over a continuous axis, typically along the x-axis. It is useful for visualizing how different components contribute to a whole and how they change over time or across categories.

***Data type:*** Area plots are most commonly used to represent time series data or data with a continuous dimension (e.g., time, age) on the x-axis. Each dataset represents a category, and the y-axis usually represents quantities or proportions.

***Example:*** Let's create an area plot to visualize the monthly revenue contributions of three product categories (Electronics, Clothing, and vehicles) for a retail business over a year:

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
electronics_revenue = [5000, 5200, 5300, 5500, 5600, 5800, 5900, 6000, 6200, 6300, 6400, 6500]
clothing_revenue = [3000, 3100, 3200, 3300, 3400, 3500, 3600, 3700, 3800, 3900, 4000, 4100]
vehicles_revenue = [2000, 2100, 2200, 2300, 2400, 2500, 2600, 2700, 2800, 2900, 3000, 3100]
plt.stackplot(months,electronics_revenue,clothing_revenue,vehicles_revenue,labels=["electronics","clothing","vehicles"])
plt.xlabel('Months')
plt.ylabel('Monthly Revenue')
plt.title('Monthly Revenue by Product Category')
plt.legend(loc='upper left')
plt.show()
```

### Stacked Bar Chart
A stacked bar chart is a data visualization that displays data in rectangular bars, with each bar representing a category or a data point. In a stacked bar chart, multiple data series or components are stacked on top of each other within each bar, making it easy to see the composition of each category and compare the total values. Stacked bar charts are effective for showing the part-to-whole relationships within categories.

***Data type:*** Stacked bar charts are suitable for categorical data where you want to show how different components contribute to a whole in each category. The categories are typically displayed along the x-axis, and the heights of the bars represent the total values, with segments of each bar indicating the contributions of individual components.

***Example:*** Let's create a stacked bar chart to visualize the distribution of sales revenue for a company in different regions (North, South, East, West) over a year:

```python
regions = ['North', 'South', 'East', 'West']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue_data = [
    [1000, 1200, 1300, 1100, 1400, 1500, 1600, 1400, 1300, 1100, 1200, 1000],
    [800, 900, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1500, 1400, 1300],
    [1200, 1300, 1400, 1600, 1500, 1400, 1200, 1100, 1300, 1500, 1600, 1200],
    [900, 1000, 1100, 1200, 1300, 1400, 1500, 1400, 1300, 1200, 1100, 1000]
]
colors = ['#023047', '#ffb703', '#219ebc', '#fb8500']
plt.bar(months,revenue_data[0],label=regions[0],color=colors[0])
for i in range(1,4):
    plt.bar(months,revenue_data[i],label=regions[i],bottom=revenue_data[i-1],color=colors[i])
    plt.xlabel('Months')
plt.ylabel('Monthly Revenue')
plt.title('Monthly Revenue by Region')
plt.legend(loc='upper left')
plt.show()
```

### Radar Chart  (Spider Chart or Web Chart)
A radar chart is a data visualization technique used to display multivariate data in the form of a two-dimensional chart with multiple axes radiating from a central point, resembling a spider's web. Each axis represents a different variable or dimension, and data points are plotted along these axes to show how each variable contributes to a whole. Radar charts are useful for comparing the performance or characteristics of different entities across multiple dimensions.

***Data type:*** Radar charts are suitable for visualizing multivariate data, where you have multiple variables or attributes for each entity. The variables should be continuous and typically represented on a common scale.

***Example:*** let's compare two job candidates' skills :

```python
candidates = ['Alice', 'Bob']
skills = ['Python', 'JavaScript', 'Problem Solving', 'Teamwork']
alice_ratings = [9, 5, 8, 7]
bob_ratings = [7, 8, 6, 9]

num_skills = len(skills)
angles = np.linspace(0, 2 * np.pi, num_skills, endpoint=False).tolist()



```

```python
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

ax.fill(angles, alice_ratings, 'b', alpha=0.1, label='Alice')
ax.fill(angles, bob_ratings, 'r', alpha=0.1, label='Bob')

ax.set_xticks(angles)
ax.set_xticklabels(skills)

plt.title('Skill Assessment Radar Chart')
plt.legend(loc='upper right')

plt.show()



```

### 3D Plot
A 3D plot, or three-dimensional plot, is a graphical representation of data that displays information in a three-dimensional space, typically using three axes (x, y, and z). It is used when data involves three variables or dimensions, allowing for the visualization of complex relationships or patterns in data that cannot be easily represented in two dimensions. 3D plots are commonly used in fields such as engineering, physics, geology, and data visualization.

***Data type:*** 3D plots are suitable for representing data with three variables, where each variable corresponds to one of the three spatial dimensions. This data can be continuous or discrete.

***Example:*** we'll create a 3D surface plot of a mathematical function, specifically a paraboloid :

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
fig.colorbar(surf)
plt.show()
```

### Sankey Diagram
A Sankey diagram is a type of data visualization that represents the flow of resources, energy, or information between multiple entities or categories. It is particularly useful for illustrating the distribution or transformation of data across various stages or processes. Sankey diagrams consist of nodes (representing categories or entities) and directed links (representing the flow between them), with the width of the links proportional to the quantity being transferred.

***Data type:*** Sankey diagrams work with data that involves the movement, allocation, or transformation of quantities from one category to another. This data is often presented in a tabular format with source and target categories and the corresponding flow quantities.

***Example:*** Let's consider an example of a Sankey diagram that visualizes the energy flow in a simple energy system, such as a power plant:

```python
from matplotlib.sankey import Sankey

# Define the flows in the Sankey diagram
flows = [0.25, 0.15, 0.6, -0.2, -0.1, -0.4]

# Create a Sankey diagram
fig, ax = plt.subplots()
sankey = Sankey(ax=ax, unit=None)
sankey.add(flows=flows, labels=['Input A', 'Input B', 'Input C', 'Output X', 'Output Y', 'Output Z'],
           orientations=[1, 1, 0, -1, -1, 0])
diagrams = sankey.finish()

# Customize the diagram
diagrams[0].patch.set_facecolor('#37c959')  # Set the color of the flows
plt.title('Sankey Diagram Example')
plt.show()
```
